```python
## first we import Libraries
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,21)
x
```

```python
y=2*x
y
```

```python
## line plot
plt.plot(x,y)
plt.show()
```

```python
## adding title and lable's
plt.plot(x,y)
plt.title('line plot')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show()
```

```python
## Changing line Aesthetics
plt.plot(x,y,color='orange',linestyle=':',linewidth=7)
plt.title(' first line plot')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show()
```

## adding two lines in the same plot

```python
x
```

```python
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='orange',linestyle=':',linewidth=7)
plt.plot(x,y2,color='green',linestyle='-',linewidth=5)
plt.title('line plot')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show()
```

## adding two grid plot in the same plot

```python
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='orange',linestyle=':',linewidth=7)
plt.plot(x,y2,color='green',linestyle='-',linewidth=5)
plt.title('line plot')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.grid()
plt.show()
```

## Adding sub-plots


```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='orange',linestyle=':',linewidth=7)
plt.title('line plot')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.grid()

plt.subplot(1,2,2)
plt.plot(x,y2,color='green',linestyle='-',linewidth=5)
plt.title('line plot',size=17)
plt.xlabel('x-label',size=15)
plt.ylabel('y-label',size=20)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color='orange',linestyle=':',linewidth=7)
plt.title('line plot')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.grid()

plt.subplot(2,1,2)
plt.plot(x,y2,color='green',linestyle='-',linewidth=5)
plt.title('line plot')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.grid()
plt.tight_layout()
plt.show()
```

## Bar plot

```python
student={'Ram':45,'sam':97,'kam':75}
names=list(student.keys())
values = list(student.values())
plt.title('this is bar plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.bar(names,values)
plt.show()
```

## Useing Bar plot and grid plot two gether

```python
student={'Ram':45,'sam':95,'kam':75}
names=list(student.keys())
values = list(student.values())
plt.title('marks of students')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.bar(names,values)
plt.grid()
plt.show()
```

## Horizontal Bar plot

```python
plt.title('this is bar plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.barh(names,values,color='red')  ## barh use for Horizontal plot
plt.grid()
plt.show()
```

## DoughNut-chart

```python
import matplotlib.pyplot as plt
  
# Setting labels for items in Chart
Employee = ['Roshni', 'Shyam', 'Priyanshi',
            'Harshit', 'Anmol']
  
# Setting size in Chart based on 
# given values
Salary = [40000, 50000, 70000, 54000, 44000]
  
# colors
colors = ['#FF0000', '#0000FF', '#FFFF00', 
          '#ADFF2F', '#FFA500']
# explosion
explode = (0.05, 0.05, 0.05, 0.05, 0.05)
  
# Pie Chart
plt.pie(Salary, colors=colors, labels=Employee,
        autopct='%1.1f%%', pctdistance=0.85,
        explode=explode)
  
# draw circle
centre_circle = plt.Circle((0, 00), 0.70, fc='white')
fig = plt.gcf()
  
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
  
# Adding Title of chart
plt.title('Employee Salary Details')
  
# Displaying Chart
plt.show()
```
